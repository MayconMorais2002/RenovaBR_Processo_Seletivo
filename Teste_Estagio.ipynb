# junção das duas tabelas

```python
import pandas as pd
import dask.dataframe as dd
from dask.diagnostics import ProgressBar

# Carregar os DataFrames originais (df1 e df2)
chunk_size = 10000  # Tamanho de cada pedaço

# Usar Dask para carregar e processar os DataFrames em pedaços menores
df1_dask = dd.read_csv('perfil_eleitorado_2020_filtrado.csv', usecols=['NM_MUNICIPIO','DS_FAIXA_ETARIA', 'DS_GENERO', 'DS_GRAU_ESCOLARIDADE','SG_UF'], blocksize=chunk_size)
df2_dask = dd.read_csv('SP_turno_1.csv', usecols=['NM_MUNICIPIO', 'NM_VOTAVEL', 'QT_VOTOS', 'SG_UF','QT_APTOS','NR_TURNO'], blocksize=chunk_size)

# Iterar pelos pedaços e realizar junções
merged_dfs = []
for df1_chunk, df2_chunk in zip(df1_dask.to_delayed(), df2_dask.to_delayed()):
    df1_chunk = df1_chunk.compute()
    df2_chunk = df2_chunk.compute()

    merged_df_chunk = df1_chunk.merge(df2_chunk, on=['SG_UF', 'NM_MUNICIPIO'], how='inner')
    merged_dfs.append(merged_df_chunk)

# Concatenar os resultados das junções
final_merged_df = pd.concat(merged_dfs, ignore_index=True)

# Trabalhar com o DataFrame resultante (final_merged_df)
```

# insights

```python
# Criar um DataFrame de resultados apenas com as colunas relevantes
resultados_candidatos = final_merged_df[['NM_MUNICIPIO', 'SG_UF', 'NM_VOTAVEL', 'QT_VOTOS']]

# Encontrar o candidato mais votado em cada município
candidato_mais_votado_por_municipio = resultados_candidatos.groupby(['NM_MUNICIPIO', 'SG_UF']) \
    .apply(lambda x: x.nlargest(1, 'QT_VOTOS')) \
    .reset_index(drop=True)

# Criar um ranking de candidatos mais votados
ranking_global = candidato_mais_votado_por_municipio.groupby('NM_VOTAVEL')['QT_VOTOS'].sum().reset_index()
ranking_global = ranking_global.sort_values(by='QT_VOTOS', ascending=False)

# Imprimir o ranking
print("Ranking global de candidatos mais votados:")
for index, row in ranking_global.iterrows():
    print(f"O candidato {row['NM_VOTAVEL']} recebeu um total de {row['QT_VOTOS']} votos.")
```

```python
# Nomes dos candidatos
candidato_x = 'BRUNO COVAS'  # Substitua pelo nome do candidato X
candidato_y = 'GUILHERME BOULOS'  # Substitua pelo nome do candidato Y

# Filtrar os resultados para os candidatos X e Y
resultados_candidato_x = final_merged_df[final_merged_df['NM_VOTAVEL'] == candidato_x]
resultados_candidato_y = final_merged_df[final_merged_df['NM_VOTAVEL'] == candidato_y]

# Encontrar o município onde o candidato X foi mais votado
grupo_municipio_votos_x = resultados_candidato_x.groupby('NM_MUNICIPIO')['QT_VOTOS'].sum()
municipio_mais_votado_x = grupo_municipio_votos_x.idxmax()

# Encontrar a quantidade de votos no município mais votado para o candidato X
quantidade_votos_mais_votado_x = grupo_municipio_votos_x.max()

# Encontrar o município onde o candidato Y foi mais votado
grupo_municipio_votos_y = resultados_candidato_y.groupby('NM_MUNICIPIO')['QT_VOTOS'].sum()
municipio_mais_votado_y = grupo_municipio_votos_y.idxmax()


# Encontrar a quantidade de votos no município para o candidato Y
quantidade_votos_mais_votado_y = grupo_municipio_votos_y.max()


# Imprimir os resultados para ambos os candidatos
print(f"Resultados para o candidato {candidato_x}:")
print(f"{municipio_mais_votado_x} com {quantidade_votos_mais_votado_x} votos.")

print("\n")

print(f"Resultados para o candidato {candidato_y}:")
print(f"{municipio_mais_votado_y} com {quantidade_votos_mais_votado_y} votos.")
```

```python
import pandas as pd

# Carregar o DataFrame com os resultados
resultados_df = pd.read_csv('/content/final_merged.csv')

# Lista de colunas relevantes para análise do perfil do eleitorado
colunas_perfil = ['DS_FAIXA_ETARIA', 'DS_GENERO', 'DS_GRAU_ESCOLARIDADE']

# Encontrar o perfil do eleitorado mais votante para cada candidato
perfis_mais_votados = resultados_df.groupby('NM_VOTAVEL')[colunas_perfil].agg(lambda x: x.mode().iloc[0]).reset_index()

# Imprimir os resultados
for index, row in perfis_mais_votados.iterrows():
    candidato = row['NM_VOTAVEL']
    faixa_etaria = row['DS_FAIXA_ETARIA']
    genero = row['DS_GENERO']
    grau_escolaridade = row['DS_GRAU_ESCOLARIDADE']
    print(f"Para o candidato {candidato}, o perfil do eleitorado mais votante é:")
    print(f"Faixa Etária: {faixa_etaria}")
    print(f"Gênero: {genero}")
    print(f"Grau de Escolaridade: {grau_escolaridade}")
    print("----------------------")
```
